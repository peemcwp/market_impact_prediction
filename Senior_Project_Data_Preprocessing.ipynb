```python
# import libraries
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
# Open CSV
csvdf = pd.read_csv('SEHK_2_clean.csv') 
csvdf = csvdf[csvdf['Volume'] != 0] 
csvdf
```

```python
# Set index to Date TimeSeries

csvdf['Date'] = pd.to_datetime(csvdf['Date'], dayfirst=True)
csvdf.set_index('Date', inplace=True)
```

```python
SEHK_2_df = csvdf[:'2018'].copy()
SEHK_2_df
```

```python
day_SEHK_2_df = SEHK_2_df.resample('B').last()
day_SEHK_2_df
```

```python
day_SEHK_2_df.info()
```

```python
day_SEHK_2_df.dropna(inplace=True)
```

```python
day_SEHK_2_df
```

```python
# Compute daily return
day_SEHK_2_df['Return'] = day_SEHK_2_df['Close'][1:].values / day_SEHK_2_df['Close'][:-1] - 1
```

```python
# Fill missing daily return value with previous return
day_SEHK_2_df.fillna(day_SEHK_2_df['Return'].mean(), inplace=True)
```

```python
day_SEHK_2_df
```

```python
# Compute 25-day volatility

day_SEHK_2_df['Volatility'] = day_SEHK_2_df['Return'].rolling(25).std()
day_SEHK_2_df.fillna(day_SEHK_2_df['Volatility'].mean(), inplace=True)
```

```python
day_SEHK_2_df
```

```python
day_SEHK_2_df['Date_key'] = day_SEHK_2_df.index.date
SEHK_2_df['Date_key'] = SEHK_2_df.index.date
```

```python
SEHK_2_df['Volatility'] = day_SEHK_2_df['Volatility']
SEHK_2_df.fillna(SEHK_2_df['Volatility'].mean(), inplace=True)
```

```python
day_SEHK_2_df.set_index('Date_key', inplace=True)
day_SEHK_2_df
```

```python
SEHK_2_df.set_index('Date_key', inplace=True)
SEHK_2_df
```

```python
# Merge 25-day Volatility to intraday dataset
SEHK_2_df['Volatility'] = day_SEHK_2_df['Volatility']
SEHK_2_df.fillna(SEHK_2_df['Volatility'].mean(), inplace=True)
```

```python
# day_SEHK_2_df = SEHK_2_df.resample('B').last()
# day_SEHK_2_df
SEHK_2_df
```

```python
SEHK_2_df2 = csvdf[:'2018'].copy()
day_SEHK_2_df2 = SEHK_2_df2.resample('B').mean()
day_SEHK_2_df2.dropna(inplace=True)
# day_SEHK_2_df2
daily_average_volume = day_SEHK_2_df2['Volume']
```

```python
day_SEHK_2_df2
```

```python
previous_daily_average_volume = daily_average_volume.shift(1)
```

```python
previous_daily_average_volume.fillna(previous_daily_average_volume.mean(), inplace=True)
previous_daily_average_volume
```

```python
daily_average_volume
```

```python
SEHK_2_df['Daily_Average_Volume'] = daily_average_volume.round(0).fillna(0)
```

```python
SEHK_2_df['Previous_Daily_Average_Volume'] = previous_daily_average_volume.round(0).fillna(0)
```

```python
SEHK_2_df['Trading Volume'] = SEHK_2_df['Volume'][:] / SEHK_2_df['Count'][:]
```

```python
# Rounding volume
SEHK_2_df['Trading Volume'] = SEHK_2_df['Trading Volume'].round(0).fillna(0)
```

```python
SEHK_2_df['Size'] = SEHK_2_df['Trading Volume'][:] / SEHK_2_df['Previous_Daily_Average_Volume'][:]
```

```python
SEHK_2_df['Volatility'] = SEHK_2_df['Volatility'].round(4).fillna(0)
SEHK_2_df['Size'] = SEHK_2_df['Size'].round(4).fillna(0)
```

```python
SEHK_2_df['POV'] = SEHK_2_df['Trading Volume'][:] / SEHK_2_df['Trading Volume'].rolling(30).sum()
```

```python
SEHK_2_df['POV'].fillna(SEHK_2_df['POV'].mean(), inplace=True)
```

```python
SEHK_2_df['Cost'] = (np.log(SEHK_2_df['VWAP'][:]) - (np.log((SEHK_2_df['High'][:] + SEHK_2_df['Low'][:]) / 2))).abs()*10000
SEHK_2_df['Cost'] = SEHK_2_df['Cost'].round(4)
```

```python
SEHK_2_df
```

```python
SEHK_2_df[SEHK_2_df['Cost'] != 0].count()
```

```python
SEHK_2_data = SEHK_2_df[['Size', 'Volatility', 'POV', 'Cost']]
cp_SEHK_2_data = SEHK_2_data.copy()
SEHK_2_data
```
