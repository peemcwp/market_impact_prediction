```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error

def SVR_score_low(X_train, X_test, y_train, y_test):
    model = SVR(C=50.0, kernel="linear", epsilon=1.0)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return [mean_squared_error(y_test, preds), mean_absolute_error(y_test, preds)]
    
```

```python
import matplotlib.pyplot as plt

model_SVR = SVR(C=50.0, epsilon=1.0, kernel="linear")
model_SVR.fit(X_train_low, y_train_low)
preds = model_SVR.predict(X_test_low)

test_val = y_test_low
preds_val = preds
ar = np.arange(len(y_test_low))
plt.plot(ar, np.zeros_like(ar) + test_val, 'o', c='r', label='Test data')
plt.plot(ar, np.zeros_like(ar) + preds_val, c='b', label='Predict data (SVR)')

SVR_MSE_low = SVR_score_low(X_train_low, X_test_low, y_train_low, y_test_low)[0]
SVR_MAE_low = SVR_score_low(X_train_low, X_test_low, y_train_low, y_test_low)[1]
plt.plot()
plt.show()
print("=======================================================")
print("SVR model MSE of random predictions on low volatility stocks: {}".format(SVR_MSE_low))
print("SVR model MAE of random predictions on low volatility stocks: {}".format(SVR_MSA_low))
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error

def SVR_score(X_train, X_test, y_train, y_test):
    model = SVR(C=30.0, kernel="rbf", epsilon=0.3)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return [mean_squared_error(y_test, preds), mean_absolute_error(y_test, preds)]
    
```

```python
import matplotlib.pyplot as plt

model_SVR = SVR(C=30.0, epsilon=0.3, kernel="rbf")
model_SVR.fit(X_train, y_train)
preds = model_SVR.predict(X_test)

test_val = y_test
preds_val = preds
ar = np.arange(len(y_test))
plt.plot(ar, np.zeros_like(ar) + test_val, c='g', label='Test data')
plt.plot(ar, np.zeros_like(ar) + preds_val, c='b', label='Predict data (SVR)')
plt.plot()
plt.show()
print("=======================================================")
print("SVR model MSE of random predictions on 26 different SEHK: {}".format(SVR_score(X_train, X_test, y_train, y_test)[0]))
print("SVR model MAE of random predictions on 26 different SEHK: {}".format(SVR_score(X_train, X_test, y_train, y_test)[1]))
```

```python
from lmfit import Model
```

```python
def i_star_model(size, volatility, pov, b1, a1, a2, a3, a4):
    return ((b1 * (a1 * np.power(size, a2) * np.power(volatility, a3))) * np.power(pov, a4)) + ((1 - b1)*(a1 * np.power(size, a2) * np.power(volatility, a3)))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
def istar_score(X_test, y_test):
    preds = i_star_model(X_test['Size'], X_test['Volatility'], X_test['POV'], b1, a1, a2, a3, a4)
    return [mean_squared_error(y_test, preds), mean_absolute_error(y_test, preds)]
    
```

```python
y = y_train_low
size = X_train_low['Size']
volatility = X_train_low['Volatility']
pov = X_train_low['POV']

istar_model = Model(i_star_model, independent_vars=['size', 'volatility', 'pov'],  param_names=['b1', 'a1', 'a2', 'a3', 'a4'])
result = istar_model.fit(y, size=size, volatility=volatility, pov=pov, b1=0.7, a1=10, a2=0.1, a3=0.1, a4=0.1)
print(result.fit_report())

a1 = result.best_values['a1']
a2 = result.best_values['a2']
a3 = result.best_values['a3']
a4 = result.best_values['a4']
b1 = result.best_values['b1']

preds = i_star_model(X_test_low['Size'], X_test_low['Volatility'], X_test_low['POV'], b1, a1, a2, a3, a4)

test_val = y_test_low
preds_val = preds
ar = np.arange(len(y_test_low))
plt.plot(ar, np.zeros_like(ar) + test_val, 'o', c='r', label='Test data')
plt.plot(ar, np.zeros_like(ar) + preds_val, c='b', label='Predict data (I-Star)')

iStar_MSE_low = istar_score(X_test_low, y_test_low)[0]
iStar_MAE_low = istar_score(X_test_low, y_test_low)[1]

plt.plot()
plt.show()
print("====================================================")
print("I-Star model MSE of random predictions on low volatility stocks : {}".format(iStar_MSE_low))
print("I-Star model MAE of random predictions on low volatility stocks : {}".format(iStar_MAE_low))
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel

def GPR_score(X_train, X_test, y_train, y_test):
#     kernel = DotProduct() + WhiteKernel()
    model = GaussianProcessRegressor(random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    preds[preds < 0] = 0
    
    return [mean_squared_error(y_test, preds), mean_absolute_error(y_test, preds)]
    
```

```python
# kernel = DotProduct() + WhiteKernel()
model_GPR = GaussianProcessRegressor(random_state=0)
model_GPR.fit(X_train_low, y_train_low)
preds = model_GPR.predict(X_test_low)

test_val = y_test_low
preds_val = preds
ar = np.arange(len(y_test_low))
plt.plot(ar, np.zeros_like(ar) + test_val, 'o', c='r', label='Test data')
plt.plot(ar, np.zeros_like(ar) + preds_val, c='b', label='Predict data (GPR)')
plt.plot()

GPR_MSE_low = GPR_score(X_train_low, X_test_low, y_train_low, y_test_low)[0]
GPR_MAE_low = GPR_score(X_train_low, X_test_low, y_train_low, y_test_low)[1]
plt.show()
print("===============================================")
print("GPR model predicting on low volatility stocks MSE: {}".format(GPR_MSE_low))
print("GPR model prediction on low volatility stocks MAE: {}".format(GPR_MAE_low))
```
