```python
X_train_low = pd.concat([X1_train, X2_train, X3_train, X7_train, X15_train, X16_train, X20_train, X22_train])
X_test_low = pd.concat([X1_test, X2_test, X3_test, X7_test, X15_test, X16_test, X20_test, X22_test])

y_train_low = pd.concat([y1_train, y2_train, y3_train, y7_train, y15_train, y16_train, y20_train, y22_train])
y_test_low = pd.concat([y1_test, y2_test, y3_test, y7_test, y15_test, y16_test, y20_test, y22_test])
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

np.random.seed(0)
y = y_train_low
X = X_train_low
parameters = {'kernel': ('linear', 'rbf','poly'), 'C':[10, 30, 40, 50, 60, 70, 100], 'epsilon':[0.1, 0.3, 0.5, 0.7, 1.0]}
svr = SVR()
clf = GridSearchCV(svr, parameters, cv=10)
clf.fit(X,y)
clf.best_params_
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error

def SVR_score(X_train, X_test, y_train, y_test):
    model = SVR(C=30.0, kernel="linear", epsilon=0.7)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return [mean_squared_error(y_test, preds), mean_absolute_error(y_test, preds)]
    
```

```python
import matplotlib.pyplot as plt

model_SVR = SVR(C=30.0, epsilon=0.7, kernel="linear")
model_SVR.fit(X_train_low, y_train_low)
preds = model_SVR.predict(X_test_low)

test_val = y_test_low
preds_val = preds
ar = np.arange(len(y_test_low))
plt.plot(ar, np.zeros_like(ar) + test_val, c='g', label='Test data')
plt.plot(ar, np.zeros_like(ar) + preds_val, c='b', label='Predict data (SVR)')
plt.plot()
plt.show()
print("=======================================================")
print("SVR model MSE of random predictions on low volatility stocks: {}".format(SVR_score(X_train_low, X_test_low, y_train_low, y_test_low)[0]))
print("SVR model MAE of random predictions on low volatility stocks: {}".format(SVR_score(X_train_low, X_test_low, y_train_low, y_test_low)[1]))
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error

def SVR_score(X_train, X_test, y_train, y_test):
    model = SVR(C=30.0, kernel="rbf", epsilon=0.3)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return [mean_squared_error(y_test, preds), mean_absolute_error(y_test, preds)]
    
```

```python
from lmfit import Model
```

```python
def i_star_model(size, volatility, pov, b1, a1, a2, a3, a4):
    return ((b1 * (a1 * np.power(size, a2) * np.power(volatility, a3))) * np.power(pov, a4)) + ((1 - b1)*(a1 * np.power(size, a2) * np.power(volatility, a3)))
```

```python
y = y_train
size = X_train['Size']
volatility = X_train['Volatility']
pov = X_train['POV']

istar_model = Model(i_star_model, independent_vars=['size', 'volatility', 'pov'],  param_names=['b1', 'a1', 'a2', 'a3', 'a4'])
result = istar_model.fit(y, size=size, volatility=volatility, pov=pov, b1=0.7, a1=10, a2=0.1, a3=0.1, a4=0.1)
print(result.fit_report())
```

```python
a1 = result.best_values['a1']
a2 = result.best_values['a2']
a3 = result.best_values['a3']
a4 = result.best_values['a4']
b1 = result.best_values['b1']

print(a1)
print(a2)
print(a3)
print(a4)
print(b1)
```

```python
def istar_score(X_test, y_test):
    preds = i_star_model(X_test['Size'], X_test['Volatility'], X_test['POV'], b1, a1, a2, a3, a4)
    return mean_squared_error(y_test, preds)
    
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor

def GPR_score(X_train, X_test, y_train, y_test):
    model = GaussianProcessRegressor(random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return mean_squared_error(y_test, preds)
    
```

```python
# SEHK 26 stocks predict result

model_GPR = GaussianProcessRegressor(random_state=0)
model_GPR.fit(X_train, y_train)
preds = model_GPR.predict(X_test)

test_val = y_test
preds_val = preds
ar = np.arange(len(y_test))
plt.plot(ar, np.zeros_like(ar) + test_val, c='g', label='Test data')
plt.plot(ar, np.zeros_like(ar) + preds_val, c='b', label='Predict data (GPR)')
plt.plot()
plt.show()
print("===============================================")
print("GPR model MAE: {}".format(GPR_score(X_train, X_test, y_train, y_test)))
```
